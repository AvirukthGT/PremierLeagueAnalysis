```python
import pandas as pd
import matplotlib.pyplot as plt
from selenium import webdriver
pd.set_option('display.max_columns',200)
```

```python
# Start headless Chrome
options = webdriver.ChromeOptions()
options.add_argument('--headless')
driver = webdriver.Chrome(options=options)

# Load page
driver.get('https://fbref.com/en/comps/9/stats/Premier-League-Stats')

# Grab rendered HTML
html = driver.page_source

# Use pandas to parse
tables = pd.read_html(html)
print(len(tables))  # see how many tables you get

# Close browser
driver.quit()


```

```python
column_rename_map = {
    # Core Info
    'Rk': 'Rank',
    'Player': 'Player',
    'Nation': 'Nationality',
    'Pos': 'Position',
    'Squad': 'Club',
    'Age': 'Age',
    'Born': 'BirthYear',

    # Playing Time
    'MP': 'MatchesPlayed',
    'Starts': 'MatchesStarted',
    'Min': 'MinutesPlayed',
    '90s': 'MatchesPlayed_90s',

    # Performance
    'Gls': 'Goals',
    'Ast': 'Assists',
    'G+A': 'GoalsPlusAssists',
    'G-PK': 'NonPenaltyGoals',
    'PK': 'PenaltyGoals',
    'PKatt': 'PenaltyAttempts',
    'G+A-PK': 'G+A_ExclPK',
    'CrdY': 'YellowCards',
    'CrdR': 'RedCards',

    # Expected
    'xG': 'ExpectedGoals',
    'npxG': 'NonPenalty_xG',
    'xAG': 'ExpectedAssists',
    'xG+xAG': 'xG_plus_xAG',
    'npxG+xAG': 'NonPenalty_xG_plus_xAG',

    # Progression
    'PrgC': 'ProgressiveCarries',
    'PrgP': 'ProgressivePasses',
    'PrgR': 'ProgressivePassesReceived',

    # Per 90 mins (if repeated)
    'Gls_1': 'Goals_per90',
    'Ast_1': 'Assists_per90',
    'xG_1': 'xG_per90',
    'xAG_1': 'xAG_per90',
    'xG+xAG_1': 'xG_plus_xAG_per90',
    'npxG_1': 'NonPenalty_xG_per90',
    'npxG+xAG_1': 'NonPenalty_xG_plus_xAG_per90',
}
```

```python
df = df.rename(columns=column_rename_map)
```

```python
df.head()
```

```python
df.columns
```

```python
from collections import defaultdict

# Store how many times we've seen each column
col_tracker = defaultdict(int)
new_cols = []

for col in df.columns:
    col_tracker[col] += 1
    if col_tracker[col] == 1:
        new_cols.append(col)
    else:
        new_cols.append(f"{col}_per90")

df.columns = new_cols
```
